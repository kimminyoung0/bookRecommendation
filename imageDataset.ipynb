```python
import os
from PIL import Image
import shutil
folder_path = './image'
#test_path = './test_image'

def check_image(img_path):
    try:
        img = Image.open(img_path)
        width, height = img.size
        pixels = img.load()
        
        if width == 1 and height == 1: #빈 파일
            return False
        
        for y in range(height):
            for x in range(width):
                if pixels[x, y] != (255, 255, 255): #흰색이 아닌 경우가 한 번이라도 있다면 정상파일
                    return True
        
        return False
    
    except Exception as e:
        print("Error:", e)
        return False


os.makedirs(f'./blank_image', exist_ok=True)
cnt = 0
# 폴더 안의 이미지 중 책 표지가 아닌 빈 파일 blank_image folder로 이동시키기
for image_name in os.listdir(folder_path):
    image_path = os.path.join(folder_path, image_name)
    label = check_image(image_path)
    if label == True: #정상 파일
        cnt += 1
        #정상파일은 냅두고
    elif label == False:
        print("빈 파일", image_name)
        #빈 파일 blank folder로 이동.
        shutil.copy(image_path, os.path.join('./blank_image', image_name))
        os.remove(image_path)
        
print(f"총{cnt}장의 이미지 활용 가능") #총 가용 이미지 개수
```
